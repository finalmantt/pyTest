```
def longeststring(string):
    num = ['1','2','3','4','5','6','7','8','9','0']
    maxx = 0
    for i,x in enumerate(string):
        if " " in x:
            x = x.replace(" ", "")
        for j in num:
            if j in x:
                x = x.replace(j, "")
        a = len(x)
        if a > maxx :
            maxx = a
    return maxx

string = ['hello 1234566723','world 23 thailand','i am very happy 1233 ']
n = longeststring(string)
print(n)
            
```
